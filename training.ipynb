```python
import numpy as np
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('regs.csv')
```

```python
df.head(10)
```

```python
#Delete rows with result = Distinction
df1 = df.loc[df['final_result'] != "Distinction"]
```

```python
df = df1
df.shape
```

```python
# Transform the data
# Transform non-numerical into numerical col
from sklearn.preprocessing import LabelEncoder

for column in df.columns:
  if df[column].dtype == np.float64:
    continue
  df[column] = LabelEncoder().fit_transform(df[column])
```

```python
df
```

```python
df['Result'] = df['final_result']
```

```python
df = df.drop('final_result', axis = 1)
```

```python
df
```

```python
df.shape[1]
```

```python
#spit the dataset
x = df.iloc[:, 15:20].values
x.shape
```

```python
y=df.iloc[:, 20].values
```

```python
#Split the dataset into 75% training and 25% testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
#Use the random forest classifier model
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10 , criterion = 'entropy', random_state = 0)
forest.fit(x_train, y_train)
```

```python
# Get the accuracy on the training dataset
forest.score(x_train, y_train)
```

```python
#Show the confusion matrix and accuracy score for the model on the test data
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, forest.predict(x_test))

TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]

print(cm)
a1_score = ((TP + TN) / (TP +TN +FN + FP)  ) *100
print(a1_score)
```
